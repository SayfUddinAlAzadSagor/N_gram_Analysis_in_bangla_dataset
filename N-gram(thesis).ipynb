```python
import nltk
import re
import string
import csv
```

```python
f =  open("questionv2.txt", encoding="utf-8")
raw = f.read()
tokens = nltk.word_tokenize(raw)
tokens = [x for x in tokens if not re.fullmatch('[' + string.punctuation + ']+'+'?', x)]
```

```python
file = csv.writer(open('word_frequencies1.csv', 'w'))
fdist1 = nltk.FreqDist(tokens)
for k,v in fdist1.most_common():
    print (k,v)
    file.writerow([k,v])
```

```python
file = csv.writer(open('word_frequencies2.csv', 'w'))
#Create your bigrams
bgs1 = nltk.bigrams(tokens)
#compute frequency distribution for all the bigrams in the text
fdist2 = nltk.FreqDist(bgs1)
for k,v in fdist2.most_common():
    print (k,v)
    l = ' '.join(map(str,k))
    file.writerow([l,v])
```

```python
file = csv.writer(open('word_frequencies3.csv', 'w'))
#Create your bigrams
bgs2 = nltk.trigrams(tokens)
#compute frequency distribution for all the trirams in the text    
fdist3 = nltk.FreqDist(bgs2)
for k,v in fdist3.most_common():
    print (k,v)
    l = ' '.join(map(str,k))
    file.writerow([l,v])
```
